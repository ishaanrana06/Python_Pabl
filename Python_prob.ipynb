```python
4#1 Reverse an Array (In-Place)
arr = list(map(int, input("Enter elements separated by space: ").split()))

left = 0
right = len(arr) - 1

while left < right:
    arr[left], arr[right] = arr[right], arr[left]
    left += 1
    right -= 1

print("Reversed array:", arr)
```

```python
#2 Find Minimum and Maximum
arr = list(map(int, input("Enter elements separated by space: ").split()))

minimum = maximum = arr[0]

for num in arr:
    if num < minimum:
        minimum = num
    if num > maximum:
        maximum = num

print("Minimum:", minimum)
print("Maximum:", maximum)
```

```python
#3 Kth Smallest Element
arr = list(map(int, input("Enter elements separated by space: ").split()))
k = int(input("Enter k: "))

arr.sort()

print(f"{k}th smallest element:", arr[k-1])
```

```python
#4 Union of Two Arrays
a = list(map(int, input("Enter first array elements: ").split()))
b = list(map(int, input("Enter second array elements: ").split()))

result = list(set(a) | set(b))

print("Union of arrays:", sorted(result))
```

```python
#5 Largest Element in Array
arr = list(map(int, input("Enter elements separated by space: ").split()))

largest = arr[0]

for num in arr:
    if num > largest:
        largest = num

print("Largest element:", largest)
```

```python
#6 Rotate Array by One (Clockwise)
arr = list(map(int, input("Enter elements separated by space: ").split()))

last = arr[-1]

for i in range(len(arr)-1, 0, -1):
    arr[i] = arr[i-1]

arr[0] = last

print("Array after rotation:", arr)
```

```python
#7 Maximum Subarray Sum (Kadane)
arr = list(map(int, input("Enter elements separated by space: ").split()))

max_sum = current_sum = arr[0]

for i in range(1, len(arr)):
    current_sum = max(arr[i], current_sum + arr[i])
    max_sum = max(max_sum, current_sum)

print("Maximum subarray sum:", max_sum)
```

```python
#8 Search Insert Position
arr = list(map(int, input("Enter sorted elements separated by space: ").split()))
target = int(input("Enter target: "))

left = 0
right = len(arr) - 1

while left <= right:
    mid = (left + right) // 2
    if arr[mid] == target:
        print("Position:", mid)
        break
    elif arr[mid] < target:
        left = mid + 1
    else:
        right = mid - 1
else:
    print("Position:", left)
```

```python
#9 Two Sum
arr = list(map(int, input("Enter elements separated by space: ").split()))
target = int(input("Enter target: "))

hashmap = {}

for i, num in enumerate(arr):
    complement = target - num
    if complement in hashmap:
        print("Indices:", [hashmap[complement], i])
        break
    hashmap[num] = i
```

```python
#10 Minimum Number of Jumps
arr = list(map(int, input("Enter elements separated by space: ").split()))
n = len(arr)

if n <= 1:
    print("Minimum jumps: 0")

elif arr[0] == 0:
    print("Minimum jumps: -1")

else:
    max_reach = arr[0]
    steps = arr[0]
    jumps = 1

    for i in range(1, n):
        if i == n - 1:
            print("Minimum jumps:", jumps)
            break

        max_reach = max(max_reach, i + arr[i])
        steps -= 1

        if steps == 0:
            jumps += 1
            if i >= max_reach:
                print("Minimum jumps: -1")
                break
            steps = max_reach - i
```
